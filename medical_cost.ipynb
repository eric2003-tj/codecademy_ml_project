```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("medical_cost.csv")
data.columns
```

```python
import seaborn as sns
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
numeric_features = ["age","bmi","children"]
X_num = data[numeric_features]
corr_matrix = X_num.corr(method='pearson')  # 'pearson' is default
sns.heatmap(corr_matrix, annot=True, cmap='RdBu_r')
plt.show()
```

```python
plt.scatter(data["bmi"],data["medical charges"])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
bmi_reshape = np.array(data["bmi"]).reshape(-1,1)
bmi_reshape = scaler.fit_transform(bmi_reshape)
data["bmi_reshape"] = pd.DataFrame(bmi_reshape)
```

```python
data.columns
```

```python
plt.scatter(data["bmi_reshape"],data["medical charges"])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
features = ["bmi_reshape","age"]
X = data[features]
y = data["medical charges"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33,random_state = 42,shuffle=True)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_train,y_train)
```

```python
data["smoker"] = data["smoker"].replace({"yes":1,"no":0})
features.append("smoker")
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```
